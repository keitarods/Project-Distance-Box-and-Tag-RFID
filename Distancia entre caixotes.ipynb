```python
import pandas as pd
import numpy as np
import geopy.distance as gd
import os
from os import path
import dtale
from numpy import cos, sin, arcsin, sqrt
from math import radians
```

# Declarando as funções para calculo das distâncias provindas dos prints do LOG

```python
def haversine(df):
    lon1 = df["Longitude da tag"]
    lat1 = df["Latitude da tag"]
    lon2 = df["Longitude do caixote"]
    lat2 = df["Latitude do caixote"]
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a))
    km = 6367 * c
    return km

def haversinet(df):
    lon1 = df["Longitude do caixote"]
    lat1 = df["Latitude do caixote"]
    lon2 = df["Longitude do trator"]
    lat2 = df["Latitude do trator"]
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a))
    km = 6367 * c
    return km
```
